```python
non_iid_bBN=pd.read_csv("fedPROX_['nonIID']_['balanced']_batch_norm[10]_100_lr_0.0001_optimizer_adam.csv")
non_iid_uBN=pd.read_csv("fedPROX_['nonIID']_['unbalanced']_batch_norm[10]_100_lr_0.0001_optimizer_adam.csv")

non_iid_bGN=pd.read_csv("fedPROX_['nonIID']_['balanced']_group_norm['']_100_lr_0.001_optimizer_sgd.csv")
non_iid_uGN=pd.read_csv("fedPROX_['nonIID']_['unbalanced']_group_norm['']_100_lr_0.001_optimizer_sgd.csv")
```

```python
epochs=np.linspace(1,100, 100)
fig, ax = plt.subplots(2, 2, figsize=(20, 7))



ax[0][0].plot(epochs, non_iid_bBN['train_acc'], label='BN')
ax[0][0].plot(epochs, non_iid_bGN['train_acc'],  label='GN')
ax[0][0].set_title("non-IID Balanced")

ax[0][0].set_ylabel('Accuracy')
ax[0][0].set_xlabel('# Rounds')
ax[0][0].grid()

ax[0][1].plot(epochs, non_iid_uBN['train_acc'],  label='BN')
ax[0][1].plot(epochs, non_iid_uGN['train_acc'],  label='GN')
ax[0][1].set_title("non-IID Unbalanced")
ax[0][1].set_xlabel('# Rounds')

ax[0][1].grid()



ax[1][0].plot(epochs, non_iid_bBN['train_loss'],  label='BN')
ax[1][0].plot(epochs, non_iid_bGN['train_loss'],  label='GN')
ax[1][0].grid()

ax[1][0].set_ylabel('Train Loss')
ax[1][0].set_xlabel('# Rounds')


ax[1][1].plot(epochs, non_iid_uBN['train_loss'],  label='BN')
ax[1][1].plot(epochs, non_iid_uGN['train_loss'],  label='GN')
ax[1][1].set_xlabel('# Rounds')

ax[1][1].grid()

legendEntries = ("BN","GN")
# set figure legend entries, number of columns, location
fig.legend(legendEntries,ncol=len(legendEntries),loc="lower center")
#plt.xlabel('Epoch')

plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))
plt.plot(figsize=(20, 7))
plt.plot(epochs, non_iid_bBN['train_acc'], label="non-IID balanced BN")
plt.plot(epochs, non_iid_bGN['train_acc'], label="non-IID balanced GN", linestyle = '--')
plt.plot(epochs, non_iid_uBN['train_acc'], label="non-IID unbalanced BN")
plt.plot(epochs, non_iid_uGN['train_acc'], label="non-IID unbalanced GN", linestyle = '--')


plt.xlabel("# Rounds")
plt.ylabel("Accuracy")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```
