```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
%matplotlib inline
```

```python
iidBN=pd.read_csv("fedGKT_commRound_10_['iid']_['unbalanced']_batch_norm[128]_10_lr_0.001_optimizer_adam.csv")
non_iid_bBN=pd.read_csv("fedGKT_commRound_10_['nonIID']_['balanced']_batch_norm[128]_10_lr_0.001_optimizer_adam.csv")
non_iid_uBN=pd.read_csv("fedGKT_commRound_10_['nonIID']_['unbalanced']_batch_norm[128]_10_lr_0.001_optimizer_adam.csv")

iidGN=pd.read_csv("fedGKT_commRound_10_['iid']_['unbalanced']_group_norm['']_10_lr_0.001_optimizer_sgd.csv")
non_iid_bGN=pd.read_csv("fedGKT_commRound_10_['nonIID']_['balanced']_group_norm['']_10_lr_0.001_optimizer_sgd.csv")
non_iid_uGN=pd.read_csv("fedGKT_commRound_10_['nonIID']_['unbalanced']_group_norm['']_10_lr_0.001_optimizer_sgd.csv")
```

```python
epochs=np.linspace(1,10, 10)
fig, ax = plt.subplots(2, 3, figsize=(20, 7))


ax[0][0].plot(epochs, iidBN['train_accuracy'],  label='BN')
ax[0][0].plot(epochs, iidGN['train_accuracy'],  label='GN')
ax[0][0].set_title("IID")
ax[0][0].set_ylabel('Accuracy')
ax[0][0].set_xlabel('# Rounds')
ax[0][0].grid()

# ax[0][1].plot(epochs,  non_iid_bBN['train_accuracy'],marker='s', label='BN')
ax[0][1].plot(epochs,  non_iid_bBN['train_accuracy'], label='BN')
ax[0][1].plot(epochs, non_iid_bGN['train_accuracy'],  label='GN')
ax[0][1].set_title("non-IID Balanced")
ax[0][1].set_xlabel('# Rounds')
ax[0][1].grid()

ax[0][2].plot(epochs, non_iid_uBN['train_accuracy'], label='BN')
ax[0][2].plot(epochs, non_iid_uGN['train_accuracy'], label='GN')
ax[0][2].set_title("non-IID Unbalanced")
ax[0][2].set_xlabel('# Rounds')
ax[0][2].grid()

ymax = max([axi.get_ylim()[1] for axi in ax[0]])
for axi in ax[0]:
    axi.set_ylim(0, ymax)


ax[1][0].plot(epochs, iidBN['train_loss'],  label='BN')
ax[1][0].plot(epochs, iidGN['train_loss'],  label='GN')
ax[1][0].set_ylabel('Train Loss')
ax[1][0].set_xlabel('# Rounds')
ax[1][0].grid()

ax[1][1].plot(epochs, non_iid_bBN['train_loss'], label='BN')
ax[1][1].plot(epochs, non_iid_bGN['train_loss'], label='GN')
ax[1][1].set_xlabel('# Rounds')
ax[1][1].grid()

ax[1][2].plot(epochs, non_iid_uBN['train_loss'], label='BN')
ax[1][2].plot(epochs, non_iid_uGN['train_loss'], label='GN')
ax[1][2].set_xlabel('# Rounds')
ax[1][2].grid()

ymax = max([axi.get_ylim()[1] for axi in ax[1]])
for axi in ax[1]:
    axi.set_ylim(0, ymax)

legendEntries = ("BN","GN")
# set figure legend entries, number of columns, location
fig.legend(legendEntries,ncol=len(legendEntries),loc="lower center")
#plt.xlabel('Epoch')

plt.show()
```

```python
fig = plt.figure(figsize=(13, 10))
plt.plot(figsize=(20, 7))
plt.plot(epochs, iidBN['train_accuracy'], label="IID BN")
plt.plot(epochs, iidGN['train_accuracy'], label="IID GN", linestyle = '--')
plt.plot(epochs, non_iid_bBN['train_accuracy'], label="non-IID balanced BN")
plt.plot(epochs, non_iid_bGN['train_accuracy'], label="non-IID balanced GN", linestyle = '--')
plt.plot(epochs, non_iid_uBN['train_accuracy'], label="non-IID unbalanced BN")
plt.plot(epochs, non_iid_uGN['train_accuracy'], label="non-IID unbalanced GN", linestyle = '--')


plt.xlabel("# Rounds")
plt.ylabel("Accuracy")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```
